```python
import numpy as np
import base64
from IPython.display import HTML, display

# -----------------------
#  IPython.display + Pillow (PIL)
# -----------------------
def show_raw_pil(filename, width, height):
    from PIL import Image as PILImage
    from IPython.display import display
    import numpy as np
    
    with open(filename, 'rb') as f:
        data = np.frombuffer(f.read(), dtype=np.uint8).reshape((height, width, 3))
    
    #  numpy  PIL Image
    img = PILImage.fromarray(data, 'RGB')
    
    #  display
    display(img)

# ---  ---
WIDTH = 640
HEIGHT = 496
FILENAME = r"D:\C-Codes\Radio-HAM\sstv_demod\cmake-build-debug\bin\output.raw"

#  Pillow 
show_raw_pil(FILENAME, WIDTH, HEIGHT)
```
